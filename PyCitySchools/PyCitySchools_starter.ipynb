# PyCity Schools Analysis

- Summary of Analysis: This data captures student test scores from a district made up of 15 high schools, both district and charter, serving 39,170 students. On average, 75% of students in 9-12 pass math, 86% pass reading, and 65% pass both. Cabrera High School, serving 1858 students with a budget of 582 dollars per student has over 90% of students pass both reading and math. Reodriguez High School, serving 3999 students is the lowest performing high school. RHS has a per student budget of 637 dollars, but only 53% of students pass reading and math. Test scores across grade bands are consistent in each school across 9-12 in both reading and math suggesting no grade is outperforming another. On average charter schools outperform district schools, and small/medium size schools outperform larger schools.
- Key Takeaways:
  1. Schools with lower student populations are performing better than schools with high populations. This could be due to  lower teacher to student ratios, or the nature of a samller program. All top performing schools serve ~2000 students, whereas all bottom performing schools serve ~4000. There coudld be other external factors impacting these outcomes such as the demographic and socioeconomic make up of these schools, but it is worth noticing the score discrepency in schools of different sizes.
  2. Per Student Spending does not appear to be correlated negativley or positively to student performance. CHS, the highest performing school has a very low per student budget, while RHS, the lowest performing school, has a much higher per student budget. In addressing school achievement gaps, the district may want to look for alternate solutions to providing low performing schools with more per pupil funding.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_column = school_data_complete.iloc[:, 4]
school_count = school_column.nunique()
print(f"Total Number of Unique Schools:{school_count}")
```

```python
# Calculate the total number of students
student_column = school_data_complete.iloc[:, 3]
student_count = student_column.count()

# Format the number with commas
formatted_student_count = "{:,.0f}".format(student_count)
print(f"Total Number of Students:{formatted_student_count}")
```

```python
# Calculate the total budget
total_budget = school_data_complete['budget'].drop_duplicates().sum()

# Format the number with commas and a dollar sign
formatted_total_budget = "${:,.0f}".format(total_budget)
print(f"Total Budget:{formatted_total_budget}")
```

```python
# Calculate the average (mean) math score
math_column = school_data_complete.iloc[:, 6]
average_math_score = math_column.mean()

# Format the number
formatted_math_mean = "{:,.2f}".format(average_math_score)

print(f"Average Math Score:{formatted_math_mean}")
```

```python
# Calculate the average (mean) reading score
reading_column = school_data_complete.iloc[:, 5]
average_reading_score = reading_column.mean()

# Format the number
formatted_reading_mean = "{:,.2f}".format(average_reading_score)

print(f"Average Reading Score:{formatted_reading_mean}")
```

```python
# Calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100

# Format the number
formatted_math_percentage = "{:.0%}".format(passing_math_percentage / 100).strip("'")

print(f"Percent of Students Passing Math:{formatted_math_percentage}")
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100

# Format the number
formatted_reading_percentage = "{:.0%}".format(passing_reading_percentage / 100).strip("'")

print(f"Percent of Students Passing Reading: {formatted_reading_percentage}")
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100

# Format the number
formatted_passing_rate = "{:.0%}".format(overall_passing_rate / 100).strip("'")

print(f"Percent of Students Passing Reading and Math: {formatted_passing_rate}")
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_snapshot = {'Total Schools':[school_count],'Total Students':[formatted_student_count],
                     'Total Budget':[formatted_total_budget],'Average Math Score': [formatted_math_mean],
                     'Average Reading Score': [formatted_reading_mean], '% Passing Math': [formatted_math_percentage],
                     '% Passing Reading':[formatted_reading_percentage], '% Overall Passing':[formatted_passing_rate]}

district_summary = pd.DataFrame(district_snapshot)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the type per school from school_data
school_types = school_data.set_index(["school_name"])["type"]
school_types_df=pd.DataFrame(school_types)

# Rename Columns
school_types_df=school_types_df.rename(columns={
    'type': 'School Type',
})

#Preview
school_types_df
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete.groupby('school_name')['student_name'].count().reset_index(name='student_count')

# Rename Columns
per_school_counts=per_school_counts.rename(columns={
    'student_count': 'Total Students',
})

#Preview
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school from school_data

# Calculate total budget per school
per_school_budget = school_data_complete.groupby('school_name').agg(
    total_budget=('budget', 'unique'),
).reset_index()

# Extract the unique budget values
per_school_budget['total_budget'] = per_school_budget['total_budget'].apply(lambda x: x[0])

# Calculate per capita spending
per_school_budget['per_capita_spending'] = per_school_budget['total_budget'] / per_school_counts['Total Students']

# Separate into individual DataFrames
budget_df = per_school_budget[['school_name', 'total_budget']]
per_capita_df = per_school_budget[['school_name', 'per_capita_spending']]

# Rename Columns
budget_df=budget_df.rename(columns={
    'total_budget': 'Total Budget',
})

per_capita_df=per_capita_df.rename(columns={
    'per_capita_spending': 'Per Student Budget',
})
# Print the DataFrames
print("Budget DataFrame:")
print(budget_df)

print("\nPer Capita Spending DataFrame:")
print(per_capita_df)
```

```python
# Calculate the average test scores per school
per_school_math = school_data_complete.groupby('school_name').agg(
    average_math=('math_score', 'mean'),
).reset_index()

per_school_reading = school_data_complete.groupby('school_name').agg(
    average_reading=('reading_score', 'mean'),
).reset_index()

# Merge the average scores
per_school_reading_math = pd.merge(per_school_math, per_school_reading, on='school_name')

# Rename Columns
per_school_reading_math=per_school_reading_math.rename(columns={
    'average_math': 'Average Math Score',
    'average_reading': 'Average Reading Score'
})
#Preview
per_school_reading_math
```

```python
# Calculate the number of students passing math
students_passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
school_students_passing_math = students_passing_math.groupby(["school_name"]).size().reset_index(name='passing_math')

school_students_passing_math
```

```python
# Calculate the number of students passing reading
students_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
school_students_passing_reading = students_passing_reading.groupby(["school_name"]).size().reset_index(name='passing_reading')
school_students_passing_reading
```

```python
# Calculate the number of students passing both math and reading
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size().reset_index(name='passing_both')
school_students_passing_math_and_reading 
```

```python
# Merge all the data into a single DataFrame
score_summary = per_school_counts.merge(per_school_reading_math, on='school_name')
score_summary = score_summary.merge(school_students_passing_math, on='school_name')
score_summary = score_summary.merge(school_students_passing_reading, on='school_name')
score_summary = score_summary.merge(school_students_passing_math_and_reading, on='school_name')

# Calculate passing percentages
score_summary['percent_passing_math'] = (score_summary['passing_math'] / score_summary['Total Students']) * 100
score_summary['percent_passing_reading'] = (score_summary['passing_reading'] / score_summary['Total Students']) * 100
score_summary['overall_passing_rate'] = (score_summary['passing_both'] / score_summary['Total Students']) * 100

# Drop the specified columns
score_summary = score_summary.drop(columns=['Total Students','passing_math', 'passing_reading', 'passing_both'])

# Rename Columns
score_summary=score_summary.rename(columns={
    'percent_passing_math': '% Passing Math',
    'percent_passing_reading': '% Passing Reading',
    'overall_passing_rate':'% Overall Passing'
})

#Preview   
score_summary
```

```python
#Merge data into a single dataframe
per_school_summary = school_types_df.copy()
per_school_summary = pd.merge(per_school_summary, per_school_counts, on="school_name")
per_school_summary = pd.merge(per_school_summary, budget_df, on="school_name")
per_school_summary = pd.merge(per_school_summary, per_capita_df, on="school_name")
per_school_summary = pd.merge(per_school_summary, score_summary, on="school_name")

# Formatting
per_school_summary["Total Budget"] = per_school_summary["Total Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary_df=pd.DataFrame(per_school_summary)
per_school_summary_df=per_school_summary_df.rename(columns={
    'school_name': 'School Name',})

#Display without index
per_school_summary_df.style.hide(axis='index')
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby('school_name').agg(
    average_math=('math_score', 'mean'),
).reset_index()

tenth_grader_math_scores = tenth_graders.groupby('school_name').agg(
    average_math=('math_score', 'mean'),
).reset_index()

eleventh_grader_math_scores = eleventh_graders.groupby('school_name').agg(
    average_math=('math_score', 'mean'),
).reset_index()

twelfth_grader_math_scores = twelfth_graders.groupby('school_name').agg(
    average_math=('math_score', 'mean'),
).reset_index()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = ninth_grade_math_scores.merge(
    tenth_grader_math_scores, 
    on='school_name', 
    suffixes=('_9th', '_10th')
).merge(
    eleventh_grader_math_scores, 
    on='school_name', 
    suffixes=('', '_11th')
).merge(
    twelfth_grader_math_scores, 
    on='school_name', 
    suffixes=('_11th', '_12th')
)

# Rename columns for clarity
math_scores_by_grade = math_scores_by_grade.rename(columns={
    'school_name': 'School Name',
    'average_math_9th': '9th',
    'average_math_10th': '10th',
    'average_math_11th': '11th',
    'average_math_12th': '12th'
})

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade.style.hide(axis='index')
```

## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = per_school_summary_df.copy()

# Clean the 'Per Student Budget' column
# Remove any unwanted characters (e.g., currency, commas)
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].replace({'\$': '', ',': ''}, regex=True)

# Convert 'Per Student Budget' to numeric, forcing errors to NaN
school_spending_df["Per Student Budget"] = pd.to_numeric(school_spending_df["Per Student Budget"], errors='coerce')

school_spending_df.style.hide(axis='index')
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"]= pd.cut(
    school_spending_df["Per Student Budget"],
    bins=spending_bins,
    labels=labels,
    include_lowest=True)

school_spending_df.style.hide(axis='index')
```

```python
#  Calculate averages for the desired column - included the observed paramter to avoid warning

# Calculate averages for the desired columns with observed parameter
spending_math_scores = school_spending_df.groupby(
    ["Spending Ranges (Per Student)"], observed=False)["Average Math Score"].mean()

spending_reading_scores = school_spending_df.groupby(
    ["Spending Ranges (Per Student)"], observed=False)["Average Reading Score"].mean()

spending_passing_math = school_spending_df.groupby(
    ["Spending Ranges (Per Student)"], observed=False)["% Passing Math"].mean()

spending_passing_reading = school_spending_df.groupby(
    ["Spending Ranges (Per Student)"], observed=False)["% Passing Reading"].mean()

overall_passing_spending = school_spending_df.groupby(
    ["Spending Ranges (Per Student)"], observed=False)["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.merge(spending_math_scores,spending_reading_scores, on='Spending Ranges (Per Student)')
spending_summary = pd.merge(spending_summary,spending_passing_math, on='Spending Ranges (Per Student)')
spending_summary = pd.merge(spending_summary,spending_passing_reading, on='Spending Ranges (Per Student)')
spending_summary = pd.merge(spending_summary,overall_passing_spending, on='Spending Ranges (Per Student)')

# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary_df["School Size"] = pd.cut(
    per_school_summary_df["Total Students"],
    bins=size_bins,
    labels=labels,
    include_lowest=True)

per_school_summary_df.style.hide(axis='index')
```

```python
# Calculate averages for the desired columns.
size_math_scores = per_school_summary_df.groupby(["School Size"], observed=False)["Average Math Score"].mean()
size_reading_scores = per_school_summary_df.groupby(["School Size"], observed=False)["Average Reading Score"].mean()
size_passing_math = per_school_summary_df.groupby(["School Size"], observed=False)["% Passing Math"].mean()
size_passing_reading = per_school_summary_df.groupby(["School Size"], observed=False)["% Passing Reading"].mean()
size_overall_passing = per_school_summary_df.groupby(["School Size"], observed=False)["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.merge(size_math_scores,size_reading_scores, on='School Size')
size_summary = pd.merge(size_summary,size_passing_math, on='School Size')
size_summary = pd.merge(size_summary,size_passing_reading, on='School Size')
size_summary = pd.merge(size_summary,size_overall_passing, on='School Size')

# Display results
size_summary
```
